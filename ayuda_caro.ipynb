## 1. Limpiamos datos

Primero limpiamos los datos de Claves_edomex.csv, podemos ver que tiene comillas de sobra

```r
df_claves_sucias <- read.csv('Claves_edomex.csv')
head(df_claves_sucias)
```

Con esta funcion quitaremos esas comillas

```r
quita_comillas <- function(nombre_municipio){
                        nombre_sin_comillas = substr(nombre_municipio, 2, nchar(nombre_municipio) - 1)
                        return(nombre_sin_comillas)
                  }
```

Aplicamos la funcion a las dos columnas

```r
df_claves <- mutate(df_claves_sucias, nom_mun = quita_comillas(nom_mun), CVEGEO = quita_comillas(CVEGEO))
head(df_claves)
```

Agregamos la entrada para Mexico

```r
df_claves <- rbind(df_claves, c('15', 'México'))
```

Usamos la columna nom_num como renglones

```r
rownames(df_claves) <- df_claves[,'nom_mun']
head(df_claves)
```

Veamos como podemos llamar al CVEGEO dado el nom_num

```r
df_claves['Aculco', 'CVEGEO']
```
